```python
import csv

data = []

with open('rawcounts.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:  
        count = 0;
        name = ''
        for i,r in enumerate(row):
#             print(i,r)
            
            if i==1:
                count = r;
                
                
            if i>1:
                if i == len(row)-1:
                    print(i,r);
                    name += r;
                    name = name.strip('"')
                    item = {
                        "name":name,
                        "count":count
                    }
                    data.append(item)
                else:
                    name += r + ' ';

print(data)
                
#             print(', '.join(row))
#             data.append(', '.join(row))
```
